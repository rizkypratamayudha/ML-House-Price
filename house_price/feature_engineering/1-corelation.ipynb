```python
import pandas as pd
```

```python
df = pd.read_csv('../../Data/house_price/cleaning/train_cleaned_encoded.csv')
df.sample(20)
```

```python
df.info()
df.columns
```

```python
# Total Luas Area
if all(col in df.columns for col in ['TotalBsmtSF', '1stFlrSF', '2ndFlrSF']):
    df['TotalLuasArea'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']

# usia rumah 
if all(col in df.columns for col in ['YearBuilt', 'YrSold']):
    df['UsiaRumah'] = df['YrSold'] - df['YearBuilt']

# total kamar mandi
if all(col in df.columns for col in ['FullBath', 'HalfBath', 'BsmtFullBath', 'BsmtHalfBath']):
    df['TotalKamarMandi'] = df['FullBath'] + df['HalfBath'] + df['BsmtFullBath'] + df['BsmtHalfBath']

# rasio area rumah dengan area lot
if all(col in df.columns for col in ['GrLivArea', 'LotArea']):
    df['AreaRatio'] = df['GrLivArea'] / df['LotArea']
```

```python
df[['TotalLuasArea', 'UsiaRumah', 'TotalKamarMandi', 'AreaRatio', 'SalePrice']].sample(20)
```

```python
df.to_csv('../../Data/house_price/cleaning/train_engineering.csv', index=False)
```
