```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Dataset
df = pd.read_csv("../../Data/house_price/train.csv")
df.head()
```

```python
#Explore the data
print("====== Info Data =====")
df.info()
print("====== Deskripsi Data =====")
df.describe()
```

```python
#missing values
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100  

missing_percent
```

```python
missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percent
})
missing_data = missing_data[missing_data['Missing Values'] > 0].sort_values(by='Percentage', ascending=False)
missing_data
```

VISUALISASI MISSING VALUES

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=missing_data.index, y=missing_data['Percentage'])
plt.xticks(rotation=90)
plt.title('Persentase Missing Values per Kolom')
plt.xlabel('Kolom')
plt.ylabel('Persentase (%)')
plt.show()
```

DROP COLOMN > 80%

```python
cols_to_drop = missing_data[missing_data['Percentage'] > 80].index.tolist()
df = df.drop(columns=cols_to_drop)

print("Kolom yang dihapus karena lebih dari 80% missing values:", cols_to_drop)
df.info()
```

```python
# check the percentage of missing values again
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percent
})
missing_data = missing_data[missing_data['Missing Values'] > 0].sort_values(by='Percentage', ascending=False)
missing_data
```

EXPORT AFTER MISSING VALUE

```python
# export after dropping columns with high missing values
df.to_csv("../../Data/house_price/cleaning/train_cleaned_missing_value_80.csv", index=False)
# check the percentage of missing values again
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_data_type = df.dtypes
missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percent,
    'Data Type': missing_data_type
})
missing_data = missing_data[missing_data['Missing Values'] > 0].sort_values(by='Percentage', ascending=False)
missing_data
```
