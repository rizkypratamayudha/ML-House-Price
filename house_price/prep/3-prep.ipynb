```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('../../Data/house_price/cleaning/train_cleaned_imputatuion.csv')
df.sample(20)
```

```python
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_data_type = df.dtypes

missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage': missing_percent,
    'Data Type': missing_data_type
})
missing_data = missing_data[missing_data['Missing Values'] > 0].sort_values(by='Missing Values', ascending=False)
missing_data
```

```python
# object to numeric
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
print(f"Jumlah kolom kategorikal: {len(categorical_columns)}")
print("Kolom kategorikal:", categorical_columns)
```

```python
df_encoded = df.copy()

for col in categorical_columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])

    # menampilkan mapping dari kategori ke angka
    mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    print(f"Mapping untuk kolom '{col}':")

df_encoded.sample(20)
```

```python
# exporting the encoded DataFrame to a CSV file
df_encoded.to_csv('../../Data/house_price/cleaning/train_cleaned_encoded.csv', index=False)
```
