```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Data Dummy

```python
#data dummy
tahun_pengalaman = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
gaji = np.array([1500, 2000, 2500, 3000, 3500])
```

```python
# Model
model = LinearRegression()
model.fit(tahun_pengalaman, gaji)
```

```python
#Prediksi
prediksi = model.predict(tahun_pengalaman)
```

```python
tahun_baru = np.array([6, 7, 8]).reshape(-1, 1)
prediksi_baru = model.predict(tahun_baru)
print("Prediksi gaji untuk tahun pengalaman 6, 7, 8:", prediksi_baru)
```

```python
# Plot
plt.scatter(tahun_pengalaman, gaji, color='blue', label='Data asli')
plt.plot(tahun_pengalaman, prediksi, color='red', label='Regresi')
plt.scatter(tahun_baru, prediksi_baru, color='green', label='Prediksi baru')
plt.xlabel('Tahun Pengalaman')
plt.ylabel('Gaji')
plt.legend()
plt.title('Prediksi Gaji berdasarkan Pengalaman')
plt.show()
```
